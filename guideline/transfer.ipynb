```python
import pandas as pd
#read csv file
df = pd.read_csv('INPUT-DATA.csv')
df
```

```python
#delete all the datframe where ionic radius is 2.060000e-10    
df = df[df['Ionic radius '] != 2.060000e-10]
df
#drop Ionic radius column
df = df.drop(columns=['Ionic radius '])
df
```

```python
#i want each separated file for temperature values like -10,0,10,20,30,40, 50, 60, 70 which contains frequency values, ReZ and ImgZ values
output_file = 'OUTPUT-DATA.csv'

# flatten the dataframe
df = df.melt(id_vars=['Temperature'], var_name='index', value_name='value')

# Split the values into frequency, ReZ, and ImgZ
# Extract the temperature values for separation
temperatures = [-10, 0, 10, 20, 30, 40, 50, 60, 70]

# Iterate over the temperatures and create separate files
for temp in temperatures:
    # Filter the dataframe for the current temperature
    temp_df = df[df['Temperature'] == temp]
    
    # Create the output file name
    temp_output_file = f'newfiles/temperature_{temp}.csv'
    
    # Select the desired columns
    # temp_df = temp_df[['Frequency', 'Re(Z)', 'Img(Z)']]
    
    # Save the dataframe to the output file
    temp_df.to_csv(temp_output_file, index=False)
    
    print(f"File for temperature {temp} has been created: {temp_output_file}")
```
